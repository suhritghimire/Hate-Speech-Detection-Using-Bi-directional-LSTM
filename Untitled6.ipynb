# BERT CAN HANDLE PUNCTUTATION, STOPWORDS ETC.

````python
## remove handle wala @ name
pandas_df['tweet_cleaned'] = pandas_df['tweet'].str.replace('@[A-Za-z0-9]+\s?','',regex=True)
pandas_df.head()
````

# Load dataframe from Pandas

````python
from datasets import Dataset

ds = Dataset.from_pandas(pandas_df)
ds
````

````python
train_test_valid = ds.train_test_split()

test_valid = train_test_valid['test'].train_test_split()

train_test_valid_dataset = DatasetDict({
    'train' : train_test_valid['train'],
    'test' : test_valid['test'],
    'valid' : test_valid['train']
})

dataset = train_test_valid_dataset.remove_columns(['hate_speech', 'offensive_language', 'neither', 'Unnamed: 0', 'count'])

dataset
````

# TOKENIZER

````python
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
````

````python
text = "My nameee iss Suhrrriittt"

output = tokenizer(text)

output
````

````python
tokens = tokenizer.convert_ids_to_tokens(output['input_ids'])
tokens
````

````python
print(f"Tokenized text: {tokenizer.convert_tokens_to_string(tokens)}")
````

````python
print(f"Vocab size is: {tokenizer.vocab_size}")

print(f"Model max length is: {tokenizer.model_max_length}")

print(f"Model input names are: {tokenizer.model_input_names}")
````

Vocab size: 28996: This indicates that the tokenizer has a vocabulary of 28,996 unique tokens that it can recognize and convert into numerical IDs. These tokens can be words, subwords, or special symbols.

Model max length is: 512: This means that the BERT model, for which this tokenizer is designed, can process sequences of up to 512 tokens. Any text longer than this will be truncated, and shorter texts will be padded to this length if padding='max_length' is used.

Model input names are: ['input_ids', 'token_type_ids', 'attention_mask']: These are the three types of inputs that the BERT model expects when you feed it tokenized text:

input_ids: The numerical IDs corresponding to the tokens in your text.

token_type_ids: Used to differentiate between two sentences in tasks like question-answering or natural language inference (e.g., 0 for the first sentence, 1 for the second). For single-sentence tasks, it's typically all zeros.

attention_mask: Indicates which tokens are actual content and which are padding tokens, so the model knows not to pay attention to the padding during processing.

# Now lets tokenize cleaned tweet_cleaned

````python
def tokenize_function(train_dataset):
  return tokenizer(train_dataset['tweet_cleaned'], padding='max_length', truncation=True)

tokenized_dataset = dataset.map(tokenize_function, batched=True)

train_dataset = tokenized_dataset['train']
eval_dataset = tokenized_dataset['valid']
test_dataset = tokenized_dataset['test']
````

### Scenario 2: Training with Frozen BERT Layers

In this scenario, we will freeze the pre-trained BERT encoder layers and only train the newly added classification head. This is often done to leverage the powerful feature extraction capabilities of BERT while reducing the number of trainable parameters, which can help prevent overfitting on smaller datasets and speed up training.

````python
# Load a fresh model instance for the freezing experiment
# This ensures we start with a pre-trained model, not one already fine-tuned from the previous run
from transformers import AutoModelForSequenceClassification
import torch

frozen_model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-cased",
    num_labels=3
)

# Freeze the BERT base layers
for param in frozen_model.bert.parameters():
    param.requires_grad = False

# Ensure the classifier head layers are trainable
# By default, these are new layers and will have requires_grad=True, but we explicitly confirm.
for param in frozen_model.classifier.parameters():
    param.requires_grad = True

# Verify which parameters are trainable
print("Parameters to be trained:")
for name, param in frozen_model.named_parameters():
    if param.requires_grad:
        print(name)

optimizer_frozen = torch.optim.AdamW(frozen_model.parameters(), lr=5e-5)
loss_fn_frozen = torch.nn.CrossEntropyLoss()
````

````python
from transformers import Trainer, TrainingArguments

# Training arguments for the frozen model
# We'll use the same settings as before for a fair comparison, but with the frozen_model
training_args_frozen = TrainingArguments(
    output_dir="./results_frozen", # New output directory
    eval_strategy="epoch",
    learning_rate=5e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3, # Keeping consistent with previous run for comparison
    weight_decay=00.01,
    save_strategy="epoch",
    logging_steps=50,
    report_to="none",
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
)

trainer_frozen = Trainer(
    model=frozen_model,
    args=training_args_frozen,
    train_dataset=current_train_set,
    eval_dataset=current_eval_set,
    compute_metrics=compute_metrics,
)

print("\nStarting training for the frozen model...")
trainer_frozen.train()
print("Frozen model training complete.")
````
